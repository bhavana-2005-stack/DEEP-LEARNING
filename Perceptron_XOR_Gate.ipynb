Perceptron and Gate

```python
import numpy as np
X=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([1,0,0,1])
X,y
```

```python
w=np.zeros(2)
b=0.0
learning_rate=0.1
epochs=10
print("Initial weights:",w)
print("Intial Bias:",b)
```

```python
def step_function(z):
    if z>=0:
        return 1
    else:
        return 0
```

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        # 1. Compute Weighted sum
        z = np.dot(w,X[i])+b

        # 2. Apply Activation Function
        y_pred = step_function(z)

        # 3. Compute Error
        error = y[i] - y_pred

        # 4. Update Weight
        w = w + learning_rate * error * X[i]

        # 5. Update Bias
        b = b + learning_rate * error

        print(f"Input ; {X[i]}, Predicted : {y_pred}, Actual : {y[i]}, w : {w}, b : {b}")
```

```python
print("\nFinal Weight : ",w)
print("Final Bias : ",b)
```

```python
print("\nTesting AND Gate")
for i in range(len(X)):
    z = np.dot(w,X[i])+b
    prediction = step_function(z)
    print("Input : ",X[i],",Prediction : ",prediction)
```
