Step 1: Prepare Training Data

```python
import numpy as np
X=np.array([1,2,3,4]) #study hours
y=np.array([0,0,1,1]) #Pass or fail
X,y
```

Step 2: Intialize Parameters

```python
w=0.0
b=0.0
learning_rate=1
epochs=5
print("Intial weight:",w)
print("Intial Bias:",b)
```

Step 3: Define Activation Function(Step Function)

```python
def step_function(z):
    if z>=0:
        return 1
    else:
        return 0
```

Step 4: Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch:{epoch+1}")
    for i in range(len(X)):
        # (i) Compute weighted sum
        z=w*X[i]+b
        # (ii) Activation function
        y_pred=step_function(z)
        # (iii) Compute error
        error=y[i]-y_pred
        # (iv) update weights
        w=w+learning_rate*error*X[i]
        # (v) update bias
        b=b+learning_rate*error

        print(f"Input:{X[i]},Predicted:{y_pred},Actual:{y[i]},w:{w},b:{b}")
```

Final Model Parameters

```python
print("Final weight:",w)
print("Final Bias:",b)
```

Testing the Model

```python
test_value=0.1
z=w*test_value+b
prediction=step_function(z)
print(f"For study hours:{test_value},Predicted class:{prediction}")
```
